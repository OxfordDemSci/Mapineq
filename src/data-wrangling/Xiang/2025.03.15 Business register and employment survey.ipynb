```python
def extract_to_csv(df, save_folder, nuts_year):
    # List to store extracted tables
    tables = []

    # Initialize variables to hold metadata
    current_date = None
    current_status = None
    current_measure = None
    current_table_start = 0

    # Iterate through the DataFrame row by row
    for index, row in df.iterrows():
        row_str = row.astype(str).str.strip()  # Convert to string and strip whitespace

        # Detect metadata
        if row_str[0].startswith("Date"):
            current_date = row_str[1]
        elif row_str[0].startswith("Employment status"):
            current_status = row_str[1]
        elif row_str[0].startswith("measure"):
            current_measure = row_str[1]
        elif row_str[0]=='Industry':
            current_table_start = index
        elif row_str[0].startswith('99000'):
            # If there is an existing table, save it before moving to the new one
            table = df.iloc[current_table_start:index+1].reset_index(drop=True)
            table.columns = table.iloc[0]
            table = table[2:].reset_index(drop=True)
            tables.append((current_date, current_status, current_measure, table))

    # Save each table as a separate file
    for date, status, measure, table in tables:
        if date and status and measure:
            filename = f"employment_{nuts_year}_{date}_{status}_{measure}.csv".replace(" ", "_")
            table.to_csv(save_folder + '//' + filename, index=False)
            print(f"Saved: {filename}")
```

```python
read_folder = r'C:\Users\AX\Documents\GitHub\MapIneq\src\data-wrangling\Xiang\Business Register_raw'
save_folder = r'C:\Users\AX\Documents\GitHub\MapIneq\src\data-wrangling\Xiang\Business register and employment survey'
for nuts_year in [2010, 2013, 2016]:
    raw_folder = read_folder + '\\' + 'nuts' + str(nuts_year)
    for file in tqdm(os.listdir(raw_folder)):
        read_path = read_folder + '\\' + 'nuts' + str(nuts_year) + '\\' + file
        df = pd.read_csv(read_path)
        extract_to_csv(df, save_folder, 'nuts' + str(nuts_year))
```
